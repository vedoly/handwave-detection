#PreProcessing

```python
import numpy as np
from preprocess_util import *
from visualization_util import *
wave = setAxes(np.load('data/hand_waving.npy'))
not_wave = setAxes(np.load('data/not_hand_waving.npy'))
```

```python
arm_wave = []
for sample in wave:
    new_sample = []
    for frame in sample:
        new_sample.append ( (onlyArm(frame) ) ) 
    arm_wave.append(np.array(new_sample))

arm_not_wave = []
for sample in not_wave:
    new_sample = []
    for frame in sample:
        new_sample.append ( (onlyArm(frame) ) ) 
    arm_not_wave.append(np.array(new_sample))

np.save("data/arm_wave.npy",np.array(arm_wave) )
np.save("data/arm_not_wave.npy",np.array(arm_not_wave) )
```

```python
angle_wave = extractAngles(np.array(arm_wave))
angle_not_wave = extractAngles(np.array(arm_not_wave) )
np.save("data/angle_wave.npy",angle_wave )
np.save("data/angle_not_wave.npy",angle_not_wave)
```
